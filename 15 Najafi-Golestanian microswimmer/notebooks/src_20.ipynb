This is a rerun of `src/19`, but with normal bounce-back boundary conditions.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

mpl.rc("figure", dpi=150)
mpl.rc("figure", figsize=(4,4))

plt.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
    'xtick.labelsize':15,
    'ytick.labelsize':15,
    'axes.labelsize':20,
})

import os
import re
```

```python
# extra stuff
from matplotlib.ticker import FormatStrFormatter, ScalarFormatter
import math
from scipy.spatial import cKDTree  # For finding nearest neighbors

# formatting
formatter = ScalarFormatter(useMathText=True)
formatter.set_powerlimits((-2, 2))  # Sets limits for when to use scientific notation
```

```python
# output directory
src_n = 20
# dir = f'../data.lbm/{src_n}/'
dir = f'../src/{src_n}/output.lbm/'

# all ticks are found using regular expressions
files = os.listdir(dir)
ticks = []

pattern = r'fluidTrj_(\d+)\.csv'

for file in files:
    match = re.search(pattern, file)
    if match:
        ticks.append(int(match.group(1)))
        
ticks.sort()

len(ticks)
```

# Fluid

```python
def fluidOverview(df):
    # figure preparation
    fig, axes = plt.subplots(2, 1, figsize = (9,3.))

    fig.suptitle(f"$t = {fluidDf.time.values[0]:.0f}$", fontsize=16)
    fig.subplots_adjust(right=0.875)  # Adjust the right space to make room for the colorbar


    # Create a meshgrid for plotting
    x_unique = fluidDf['coordinate_x'].unique()
    y_unique = fluidDf['coordinate_y'].unique()
    X, Y = np.meshgrid(x_unique, y_unique)

    # top plot
    upperLim = 0.003
    lowerLim = 0.
    cmap = plt.get_cmap('cividis')
    norm = plt.Normalize(lowerLim, upperLim)

    axes[0].streamplot(X, Y, 
        fluidDf.fluidVelocity_x.unstack().values.transpose(),
        fluidDf.fluidVelocity_y.unstack().values.transpose(),
        density=1.5, linewidth=0.5, color="black",
    )
    axes[0].pcolormesh(X, Y,
        np.sqrt(fluidDf.fluidVelocity_x**2 + fluidDf.fluidVelocity_y**2).unstack().values.transpose(),
        vmin=lowerLim,
        vmax=upperLim,
        cmap=cmap, alpha = 0.8
    );
    axes[0].tick_params(labelbottom=False)
    axes[0].set_ylabel("$y$")

    cbar_ax = fig.add_axes([0.9, 0.55, 0.02, 0.3])  # [left, bottom, width, height] for the colorbar
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    cbar = fig.colorbar(sm, cax=cbar_ax, alpha=0.8)
    cbar.set_label(label='$\\mathbf{u}$', fontsize=16)
    cbar.ax.tick_params(labelsize=15)

    custom_ticks = np.array([lowerLim, upperLim/2, upperLim])
    cbar.set_ticks(custom_ticks)
    cbar.set_ticklabels([f'{tick:.3f}' for tick in custom_ticks])

    # bottom plot
    upperLim = 1.01
    lowerLim = 0.99
    cmap = plt.get_cmap('seismic')
    norm = plt.Normalize(lowerLim, upperLim)
    axes[1].pcolormesh(X,Y,
        fluidDf.massDensity.unstack().values.transpose(),
        vmin=lowerLim,
        vmax=upperLim,
        cmap=cmap,
    );
    axes[1].set_xlabel("$x$")
    axes[1].set_ylabel("$y$")

    cbar_ax = fig.add_axes([0.9, 0.13, 0.02, 0.3])  # [left, bottom, width, height] for the colorbar
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    cbar = fig.colorbar(sm, cax=cbar_ax)
    cbar.set_label(label='$\\rho$', fontsize=16)
    cbar.ax.tick_params(labelsize=15)

    custom_ticks = np.array([lowerLim, 1 ,upperLim])
    cbar.set_ticks(custom_ticks)
    cbar.set_ticklabels(custom_ticks)
    cbar.set_ticklabels([f'{tick:.2f}' for tick in custom_ticks])
    
    return fig, axes
```

# mean fluid speed for every cycle of 200 μs

```python
cycleInitTicks = np.arange(0, len(ticks), 10)[:-1]
cycleInitTicks
```

```python
avgFluidVelocities = []

for tick in cycleInitTicks:
    avg = pd.read_csv(dir + f"fluidTrj_{ticks[tick]}.csv").filter(['id_x', 'id_y','coordinate_x','coordinate_y', 'fluidVelocity_x', 'fluidVelocity_y']).set_index(["id_x","id_y"]).sort_index()

    for subtick in ticks[tick+1:tick+11]:
        auxDf = pd.read_csv(dir + f"fluidTrj_{subtick}.csv").set_index(["id_x","id_y"]).sort_index()
        avg['fluidVelocity_x'] += auxDf.fluidVelocity_x
        avg['fluidVelocity_y'] += auxDf.fluidVelocity_y

    avg['fluidVelocity_x'] /= 11
    avg['fluidVelocity_y'] /= 11
    avgFluidVelocities.append(avg)
    print(tick)
```

```python
fig, axes = plt.subplots(len(avgFluidVelocities), 1, figsize = (9,2*len(avgFluidVelocities)))
fig.suptitle("$\\langle\\mathbf{u}\\rangle (x,y)$; three particles; complete cycles only", fontsize=16)
fig.subplots_adjust(right=0.875)  # Adjust the right space to make room for the colorbar

id = -1
for df in avgFluidVelocities:
    id += 1

    # Create a meshgrid for plotting
    x_unique = df['coordinate_x'].unique()
    y_unique = df['coordinate_y'].unique()
    X, Y = np.meshgrid(x_unique, y_unique)
    
    # top plot
    upperLim = 0.0003
    lowerLim = 0.
    cmap = plt.get_cmap('cividis')
    norm = plt.Normalize(lowerLim, upperLim)

    axes[id].streamplot(X, Y, 
        df.fluidVelocity_x.unstack().values.transpose(),
        df.fluidVelocity_y.unstack().values.transpose(),
        density=1.5, linewidth=0.5, color="black",
    )
    axes[id].pcolormesh(X, Y,
        np.sqrt(df.fluidVelocity_x**2 + df.fluidVelocity_y**2).unstack().values.transpose(),
        vmin=lowerLim,
        vmax=upperLim,
        cmap=cmap, alpha = 0.8
    );
    axes[id].tick_params(labelbottom=False)
    axes[id].set_ylabel("$y$")
    axes[id].set_title(f'$t \\in [{ticks[cycleInitTicks[id]]/10:.0f}, {ticks[cycleInitTicks[id]+10]/10:.0f})$')

cbar_ax = fig.add_axes([0.9, 0.25, 0.02, 0.5])  # [left, bottom, width, height] for the colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, cax=cbar_ax, alpha=0.8)
cbar.set_label(label='$\\mathbf{u}$', fontsize=16)
cbar.ax.tick_params(labelsize=15)

custom_ticks = np.array([lowerLim, upperLim/2, upperLim])
cbar.set_ticks(custom_ticks)
cbar.set_ticklabels([f'{tick:.4f}' for tick in custom_ticks])

axes[-1].set_xlabel("$x$")
plt.show()
```
